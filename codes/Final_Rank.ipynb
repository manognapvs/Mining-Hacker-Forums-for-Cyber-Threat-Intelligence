#HACKER WORDS

```python
import pandas as pd
csv_hacker = pd.read_csv("/content/drive/MyDrive/Dataset/10.csv_Part0-hack_words-2.csv", header=0)
display(csv_hacker)
gb = csv_hacker.groupby('txtAuthor')['hacker_words'].apply(list)
df = gb.to_frame()
display(df)

```

```python
def count_hackers(row):
  count = 0
  for i in row:
    i = i.replace("[", "")
    i = i.replace("]", "")
    if i != '':
      i = i.split(",")
      count += len(i)
  return count
```

```python
df["Hacker_count"] = df["hacker_words"].apply(count_hackers)
display(df)
```

Data Standardization

```python
max_val = int(df['Hacker_count'].max())
min_val = int(df['Hacker_count'].min())
li = list()
diff = max_val - min_val
for index,row in df.iterrows():
    val = float((int(row['Hacker_count']) - min_val)/diff)
    # print(type(val))
    li.append(val)

df['Data_Standardization'] = li
display(df)
```

Entropy

```python
from math import log
import pandas as pd

n = len(df)
k = 1/(log(n))
total = df['Data_Standardization'].sum()
p = list()
for index,row in df.iterrows():
    x = row["Data_Standardization"]
    p.append(float(x/total))

# print(p)

e = list()
for i in p:
    if i != 0.0:
        temp = log(i) * i
        e.append(float(temp))
    else:
        e.append(0.0)

entropy_temp = sum(e)
entropy = (-1*k)*entropy_temp

# df["Entropy"] = e
# display(df)

print(entropy)    
data = {'HackWords': entropy}
```

Weighted Summation

```python
li = list()
for index,row in df.iterrows():
    li.append(float(row["Data_Standardization"] * 0.24151045729006376))


df["Weighted_Summation_hackerwords"] = li
# df = df.sort_values(by=["Weighted_Summation_hackerwords"], ascending=False)

# df_users = df[["Weighted_Summation_hackerwords"]]
df_users = df_users.assign(Weighted_Summation_hackerwords=li)
display(df_users)


```

#KNOWLEDGE AQ

```python
import pandas as pd
csv11 = pd.read_csv("/content/drive/MyDrive/Dataset/10.csv_Part0.txt", sep="\t", header=0)
csv11.head(10)
df = csv11[["txtAuthor","txtBody_NoQuote_Clean"]]
df.head(5)


gb = df.groupby('txtAuthor')['txtBody_NoQuote_Clean'].apply(list)
# print(gb)
# print(type(gb))
df = gb.to_frame()
display(df)
#------To read knowledge acquistion as a list --------
my_file = open("/content/drive/MyDrive/Dataset/knowledge acquisition.txt", "r")
data = my_file.read()
dl = data.split("\n")
my_file.close()
print(dl)
#------To read Knowledge requesition file as a list-----
their_file = open("/content/drive/MyDrive/Dataset/knowledge requesition.txt", "r")
data = their_file.read()
# data = data.lower()
dr = data.split("\n")
my_file.close()
print(dr)

#-----counting the number of KR words--
def dr_checker(text):
  # text = ' '.join(text)
  # print(type(text))
  count = 0
  for x in dr:
    for ele in text:
      # print(ele)
      # count = count + 1
      if x.lower() in str(ele).lower():
        # print(type(text))
        count = count + 1
  return count 

#-----counting the number of K words--
def dl_checker(text):
  # text = ' '.join(text)
  # print(type(text))
  count = 0
  for x in dl:
    for ele in text:
      # print(ele)
      if x.lower() in str(ele).lower():
        count = count + 1
  return count
 
df["count_acquisition"] = df["txtBody_NoQuote_Clean"].apply(dl_checker)
df["count_requesition"] = df["txtBody_NoQuote_Clean"].apply(dr_checker)
display(df)
# df.to_csv(r'/content/drive/MyDrive/Dataset/export_dataframe.csv', index=False, header=True)
```

```python


max_val = df["count_acquisition"].max()
print(max_val)
min_val = df["count_acquisition"].min()
print(min_val)
li = list()

diff = max_val - min_val
for index,row in df.iterrows():
    # print(row["Length_of_messages"])
    val = float((int(row["count_acquisition"]) - min_val)/diff)
    # print(type(val))
    li.append(val)

df['Data_Standardization_count_aq'] = li
display(df)
```

```python
from math import log

n = len(df)
k = 1/(log(n))
total = df['Data_Standardization_count_aq'].sum()
p = list()
for index,row in df.iterrows():
    x = row["Data_Standardization_count_aq"]
    p.append(float(x/total))

# print(p)

e = list()
for i in p:
    if i != 0.0:
        temp = log(i) * i
        e.append(float((-1*k)*temp))
    else:
        e.append(0.0)
df["Entropy"] = e
display(df)
```

```python
li = list()
for index,row in df.iterrows():
    li.append(float(row["Data_Standardization_count_aq"] * 0.08118008835111033))

# df = df.sort_values(by=["Weighted_Summation_hackerwords"], ascending=False)
df_users = df_users.assign(Weighted_Summation_knowledge_aq=li)
# df_users = df[["Weighted_Summation_knowledge_aq"]]
# df_users["Weighted_Summation_knowledge_aq"] = li

display(df_users)
```
